# Reading Data

```python
def getfile(location_pair,**kwargs):                                                                         #tries to get local version and then defaults to google drive version
    gdrive=location_pair
    loc = 'https://drive.google.com/uc?export=download&id='+gdrive.split('/')[-2]
    out=pd.read_pickle(loc,**kwargs)
    return out

                                                                                                             # Read the data from the dataset file
fname=("https://drive.google.com/file/d/1PDrNKb-vcb6YH1o18EvvYEXf2a82_DJq/view?usp=share_link")
ppd_df=getfile(fname)
ppd_df.head()
```

# Measuring Skewness based on month and year

```python
ppd_df["Month"] = ppd_df["Month/Year"]%100
ppd_df["Year"] = ppd_df["Month/Year"]/100
ppd_df['Year'] = ppd_df['Year'].astype(int)
```

```python
print ("We have data for the following years: ", ppd_df["Year"].unique())                                      #Checking data based on year
print ("Number of rows of year 2010 is: ", (ppd_df['Year']==2010).sum())
print ("Number of rows of year 2011 is: ", (ppd_df['Year']==2011).sum())
print ("Number of rows of year 2012 is: ", (ppd_df['Year']==2012).sum())
print ("Number of rows of year 2013 is: ", (ppd_df['Year']==2013).sum())
print ("Number of rows of year 2014 is: ", (ppd_df['Year']==2014).sum())
print ("Number of rows of year 2015 is: ", (ppd_df['Year']==2015).sum())
print ("Number of rows of year 2016 is: ", (ppd_df['Year']==2016).sum())
print ("Number of rows of year 2017 is: ", (ppd_df['Year']==2017).sum())
print ("Number of rows of year 2018 is: ", (ppd_df['Year']==2018).sum())
print ("Number of rows of year 2019 is: ", (ppd_df['Year']==2019).sum())
print ("Number of rows of year 2020 is: ", (ppd_df['Year']==2020).sum())
print ("Number of rows of year 2021 is: ", (ppd_df['Year']==2021).sum())
print ("Number of rows of year 2022 is: ", (ppd_df['Year']==2022).sum())
print ("Number of rows of year 2023 is: ", (ppd_df['Year']==2023).sum())
```

```python
pd.set_option("display.max_rows", None)                                                            #Checking data based on month
yearcnts = ppd_df.value_counts(subset=['Year','Month'])
for i in ppd_df["Year"].unique():
  for j in ppd_df["Month"].unique():
    if i in yearcnts and j in yearcnts[i]:
      print("Count for year",i,"and month",j,"is: ",yearcnts[i][j])
    else:
      print("We don't have data for year",i,"and month",j)
print ("Clearly we have data from January 2010 to February 2022 and there is no skew, that is, we have roughly equal data for all the months")
```

Clearly we have data from January 2010 to February 2022 and there is no skew, that is, we have roughly equal data for all the months
